```python
def cmocean_to_plotly(cmap, pl_entries):
    h = 1.0/(pl_entries-1)
    pl_colorscale = []
    newcmap = cmocean.tools.crop_by_percent(cmap, 30, which='both', N=None)
    for k in range(pl_entries):
        C = list(map(np.uint8, np.array(cmap(k*h)[:3])*255))
        # https://stackoverflow.com/questions/6800481/python-map-object-is-not-subscriptable
        pl_colorscale.append([k*h, 'rgb'+str((C[0], C[1], C[2]))])

    return pl_colorscale
```

```python
cmap = cmocean.cm.curl
newcmap = cmocean.tools.crop_by_percent(cmap, 30, which='both', N=None)
curl = cmocean_to_plotly(newcmap, 200)
```

```python
df=pd.read_csv("ccn_diff.csv",na_values=[-9999])
```

```python
site_lat = df.lat_bin
site_lon = df.lon_bin
locations_name = df['CCN_1.0_cm3']- df['CCN_1.0_cm31']
l0_name = df['SS_1.0_mg']-df['SS_1.0_mg1']
```

```python
data = Data([
    go.Scattermapbox(
        # 绘制散点的经纬度
        lat=site_lat,
        lon=site_lon,
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=8,
            color=locations_name,
            showscale=True,
            cmax=750,
            cmin=-750,
            colorscale=curl
        ),
        # 散点对应的文本
        text=locations_name,
        customdata=l0_name,
        name='CCN'
    )
])
```

```python
layout = Layout(
    title='CCN',
    autosize=True,
    width=800,
    height=800,
    hovermode='closest',
    mapbox=dict(
        accesstoken=mapbox_access_token,
        domain={'x': [0, 1.0], 'y': [0, 1]},
        bearing=0,
        # 地图中心坐标，不要远离绘制的散点坐标
        center=dict(
            lat=32,
            lon=118
        ),
        pitch=0,
        zoom=4
    ),
)
```

fig = go.FigureWidget(data=data, layout=layout)
