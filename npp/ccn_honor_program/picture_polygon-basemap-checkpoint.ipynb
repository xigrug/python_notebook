```python
import pandas as pd
import numpy as np
import numpy.ma as ma
import matplotlib.pyplot as plt
from matplotlib import cm as CM
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
```

```python
%matplotlib inline
```

```python
# read location data
namesa=['LOC1','LOC2','LOC3','LOC4','LOC5','LOC6','LOC7','LOC8']
LOC=pd.read_table('location_use13-06-17.txt',sep='\s+',names=namesa,na_values=[9999])
a=LOC['LOC1'].shape
num_ploy=a[0]
# Extract the data we're interested in
latmax=LOC.iloc[:,:4].max().max()
latmin=LOC.iloc[:,:4].min().min()
lonmax=LOC.iloc[:,4:8].max().max()
lonmin=LOC.iloc[:,4:8].min().min()
llcrnrlon=lonmin-0.5
print(llcrnrlon)
llcrnrlat=latmin-0.5
print(llcrnrlat)
urcrnrlon=lonmax+0.5
print(urcrnrlon)
urcrnrlat=latmax+0.5
print(urcrnrlat)
```

```python
# 1. Draw the map background
fig = plt.figure(figsize=(8, 8))
#m = Basemap(projection='lcc', resolution='h', 
#            lat_0=31.5, lon_0=117.5,
#            width=1E6, height=1.2E6)
m = Basemap(resolution='f',llcrnrlon=llcrnrlon,llcrnrlat=llcrnrlat,
            urcrnrlon=urcrnrlon,urcrnrlat=urcrnrlat, epsg=4269)# default projection=PlateCarree
#m.arcgisimage(service='ESRI_Imagery_World_2D', xpixels = 1500, verbose= True)
#######
#m.shadedrelief()
#m.drawcoastlines(color='b')
#m.drawcountries(color='gray')
#m.drawstates(color='gray')
#add lon lat sticks
m.drawparallels(np.arange(20.,40,2.5),linewidth=1, dashes=[4, 2], labels=[1,0,0,0], color= 'gray',zorder=0, fontsize=10)
m.drawmeridians(np.arange(100.,125.,2.),linewidth=1, dashes=[4, 2], labels=[0,0,0,1], color= 'gray',zorder=0, fontsize=10)
##add shapefile
#m.readshapefile('dijishi_2004','whatevername',color='gray')
m.readshapefile('dijishi_2004','state',color='gray')
```

```python
def plot_rec(map,lower_left,upper_left,upper_right,lower_right,text):
    lon_poly = np.array([lower_left[0], upper_left[0],upper_right[0], lower_right[0]])
    lat_poly = np.array([lower_left[1], upper_left[1],upper_right[1], lower_right[1]])
    X, Y  = map(lon_poly, lat_poly)
    xy = np.vstack([X,Y]).T

    poly = Polygon(xy, closed=True, \
            facecolor='None',edgecolor='red', \
            linewidth=1.\
            )
    ax, ay = map(lower_left[0],lower_left[1])
    plt.text(ax, ay,text,fontsize=6,fontweight='bold',
                    ha='left',va='bottom',color='k')
    plt.gca().add_patch(poly)

def plot_rec3(map,lower_left,upper_left,upper_right,text):
    lon_poly = np.array([lower_left[0], upper_left[0],upper_right[0]])
    lat_poly = np.array([lower_left[1], upper_left[1],upper_right[1]])
    X, Y  = map(lon_poly, lat_poly)
    xy = np.vstack([X,Y]).T

    poly = Polygon(xy, closed=True, \
            facecolor='None',edgecolor='red', \
            linewidth=1.\
            )
    ax, ay = map(lower_left[0],lower_left[1])
    plt.text(ax, ay,text,fontsize=6,fontweight='bold',
                    ha='left',va='bottom',color='k')
    plt.gca().add_patch(poly)

for i in range(num_ploy):
    print(i)
    plot_rec(m,[LOC.iloc[i,4].round(3),LOC.iloc[i,0].round(2)],[LOC.iloc[i,5].round(3),
             LOC.iloc[i,1].round(2)],[LOC.iloc[i,6].round(3),LOC.iloc[i,2].round(2)],
             [LOC.iloc[i,7].round(3),LOC.iloc[i,3].round(2)],str(i+1))

m.imshow(plt.imread('nasa-worldview-2013-06-17-1513612712.jpg'),origin='upper')
#plt.savefig('showpolygon-130617.png',dpi=600)
plt.show()
```
