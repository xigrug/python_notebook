```python
def mlr_ccn_value(value,value1,value2,value3):
    #value_name,value1_name,value2_name,value3_name="CCN_1.0_cm3","fa1","fa2","fa3"
    value_name,value1_name,value2_name,value3_name="CCN_1.0_cm3","T2m_NCEP","SS_1.0_mg","Ttop1.0"
    data=pd.DataFrame()
    data[value_name]=value
    data[value1_name]=value1
    data[value2_name]=value2
    data[value3_name]=value3
    data.dropna()
    X = data[[value1_name,value2_name,value3_name]].values.reshape(-1,3)
    y = data[value_name]
    lm = linear_model.LinearRegression()
    model = lm.fit(X,y)
    #predictions = lm.predict(X)
    #print(predictions[0:5])
    #输出我们模型的R²得分
    print(lm.score(X,y))
    cf=lm.coef_
    inter=lm.intercept_
    return cf[0],cf[1],cf[2],inter
```
