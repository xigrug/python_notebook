```python
import pandas as pd
import glob
import datetime
```

```python
data=pd.read_csv("test.tgs")
```

```python
data.head()
```

```python
data1=data.values.reshape(720, 20*25)
```

```python
data2 =data1[::2]
```

```python
data2=data2.reshape(9000, 20)
data2=pd.DataFrame(data2,columns=["start_year","start_month","start_date","start_hour","year","month","day","hour","age_hour","latitude","longitude","height","PRESSURE", "AIR_TEMP", "RAINFALL", "MIXDEPTH", "RELHUMID", "SPCHUMID", "H2OMIXRA", "TERR_MSL"])
```

```python
data2.head()
```

```python
data2['year']=data2['year'].astype("int64")
data2['month']=data2['month'].astype("int64")
data2['day']=data2['day'].astype("int64")
data2['hour']=data2['hour'].astype("int64")
data2['start_year']=pd.to_datetime(data2['start_year'],format='%y').dt.strftime('%y')
data2['start_month']=pd.to_datetime(data2['start_month'],format='%m').dt.strftime('%m')
data2['start_date']=pd.to_datetime(data2['start_date'],format='%d').dt.strftime('%d')
data2['start_hour']=pd.to_datetime(data2['start_hour'],format='%H').dt.strftime('%H')
data2["longitude"]=data2["longitude"].round(3)
data2["latitude"]=data2["latitude"].round(3)
```

```python
data2.head()
```

```python
data2.to_csv("traj8_24_4-10_2h.tgs",index=False)
```
